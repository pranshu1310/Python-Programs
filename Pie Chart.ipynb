Q26. Write a program in python to plot a pie chart on consumption of water in daily life.

```python
import matplotlib.pyplot as plt
import random
days = [i for i in range(1,10)]
sizes = [random.randint(0,10) for i in range(1,10)]
plt.pie(sizes, labels=days, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.show()
```
