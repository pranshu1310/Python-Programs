Q15. Write a python function sin(x,n) to calculate the value of sin(x) using its Taylorseries expansion
up to n terms.



```python
import math
def sin(x,n):
    sine = 0.0
    for i in range(n):
        sign = (-1)**i
        pi=22/7
        y=x*(pi/180)
        sine = sine + ((y**(2.0*i+1))/math.factorial(2*i+1))*sign
    return sine
x=int(input("Enter the value of x in degrees:"))
n=int(input("Enter the number of terms:"))
print(round(sin(x,n),2))
```
